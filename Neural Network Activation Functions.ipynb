# Activation Functions

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x1 = np.random.randint(-10,20,10)
x2 = np.random.randint(-7,20,10)
```

```python
x1,x2
```

```python
X1, X2 = np.meshgrid(x1, x2)
X1,X2
```

```python
y = 3*X1 + 4*X2
```

# Sigmoid Function

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
z = sigmoid(y)
```

```python
z
```

```python
# Plot the sigmoid output 2d
plt.subplots(figsize = (8,5))
plt.scatter(y,z, s=40) # size of the scatter
plt.xlabel("y")
plt.ylabel("z")
plt.show()
```

```python
# Plot the sigmoid output
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X1,X2, z, cmap='viridis')
```

# tanh Activation function

```python
def tanh(x):
    return np.tanh(x)
```

```python
z1 = tanh(y)
```

```python
# Plot the tanh output 2d
plt.subplots(figsize = (8,5))
plt.scatter(y,z1, s=40) # size of the scatter
plt.xlabel("y")
plt.ylabel("z1")
plt.show()
```

```python
# Plot the tanh output
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X1,X2, z1, cmap='viridis')
```

# ReLU Activation Function

```python
def relu(x):
    return np.maximum(0, x)
```

```python
z2 = relu(y)
```

```python
# Plot the Relu output 2d
plt.subplots(figsize = (8,5))
plt.scatter(y,z2, s=40) # size of the scatter
plt.xlabel("y")
plt.ylabel("z2")
plt.show()
```

```python
# Plot the Relu output
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X1,X2, z1, cmap='viridis')
```

# 2. Sample Data

```python
import pandas as pd
```

```python
# y = 3*x1 + 4*x2
```

```python
x1 = 3
x2 = 2
y = 17 
```

```python
lr = 0.01
w1 = 3
w2 = 7
error =[]
prediction = []

for epoch in range(20):
    y_pred = x1*w1 + x2*w2
    squ_err = (y-y_pred)**2
    dEw1 = 2*(y-y_pred)*(-x1)
    dEw2 = 2*(y-y_pred)*(-x2)
    error.append(squ_err)
    prediction.append(y_pred)
    
    w1 = w1 - lr*dEw1
    w2 = w2 - lr*dEw2
    
    print("value of W1",w1,"value of w2",w2,"Error is",squ_err )
```

```python
print(w1,w2)
```

```python
x1*w1 + x2*w2
```

# Plot the progression of the error across epochs.

```python
# Plot for Error
plt.subplot(3,1,1)
plt.plot(range(len(error)), error, label='NM', color='red')
plt.title('error_vs_epoch')
plt.xlabel('Number of updates/iterations')
plt.ylabel('error')
plt.legend()
```

# Plot the progression of ypred across epochs

```python
# Plot for Error
plt.subplot(3,1,1)
plt.plot(range(len(prediction)), prediction, label='NM', color='blue')
plt.title('y_pred_vs_epoch')
plt.xlabel('Number of updates/iterations')
plt.ylabel('y_pred')
plt.legend()
```
